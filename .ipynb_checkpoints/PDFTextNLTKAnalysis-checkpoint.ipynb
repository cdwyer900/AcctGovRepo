```python
#! python3
# combinePdfs.py - Combines all the PDFs in the current working directory into 
# a single PDF.

import PyPDF2, os

# Get all the PDF filenames.
pdfFiles = []
for filename in os.listdir('.'):
    if filename.endswith('.pdf'):
        print(filename)
        pdfFiles.append(filename)
pdfFiles.sort()

pdfWriter = PyPDF2.PdfFileWriter()

# Loop through all the PDF files.
for filename in pdfFiles:
    pdfFileObj = open(filename, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

    # Loop through all the pages (except the first) and add them.
    for pageNum in range(1, pdfReader.numPages):
        pageObj = pdfReader.getPage(pageNum)
        pdfWriter.addPage(pageObj)

# Save the resulting PDF to a file.
pdfOutput = open('pdfoutput.pdf', 'wb')
pdfWriter.write(pdfOutput)
pdfOutput.close()
```

```python
# extract to text
from PyPDF2 import PdfReader

#reader = PdfReader("pdfoutput.pdf") this file also has the visa esg report

reader = PdfReader("2021-ESG-Visa.pdf")
text = ""
for page in reader.pages:
    text += page.extract_text() + "\n"
#write out a copy of the file as text
outFile=open("VISA-ESG-txt.txt","w")
outFile.write(text)
    
#text is a string
#for Visa
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.corpus import stopwords
import csv
import string

#fin = open("visa_content.txt","r")

#text_in = fin.read()

""" remove stopwords """
stop_words = set (stopwords.words('english'))
stop_words.add("Close")
stop_words.add("Modal")
stop_words.add("Navigation")
stop_words.add("navigation")
stop_words.add("Navigate")
stop_words.add("menu")
stop_words.add("Move")
stop_words.add("left")
stop_words.add("right")
stop_words.add("The")
stop_words.add("’")
stop_words.add('”')
stop_words.add('“')
stop_words.add("•")
stop_words.add("also")



result_str =  ' '.join([word for word in text.split() if word not in stop_words])
result_str2 = ''.join([word for word in result_str if word not in string.punctuation])
result_str2 = result_str2.strip("’”“")

#remove punctuation

word_tokens = word_tokenize(result_str2)
#word_tokens = word_tokenize(text_in)
text1 = nltk.Text(word_tokens)

ngram_list = nltk.ngrams(word_tokens,2)
ngram_fd = nltk.probability.FreqDist(ngram_list)
t_fd = nltk.probability.FreqDist(word_tokens)

top_ngrams = ngram_fd.most_common(100)
sorted_ngrams = sorted(top_ngrams)

out_list = []
out_list2 = []
out_str = ""

for line in top_ngrams:
    ngram, num = line
    print (ngram)
    out_list.append(ngram)
    word1, word2 = ngram
    out_list2.append([word1, word2])



```
