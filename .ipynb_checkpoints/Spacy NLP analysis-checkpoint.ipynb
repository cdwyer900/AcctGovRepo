```python
import spacy
import spacy_cleaner
from spacy_cleaner.processing import removers, replacers, mutators
from spacy.tokens import Doc
import re

model = spacy.load("en_core_web_lg")
pipeline = spacy_cleaner.Pipeline(
    model,
    removers.remove_url_token,
    #removers.remove_stopword_token,
    #removers.remove_punctuation_token,
    #replacers.replace_punctuation_token,
    #mutators.mutate_lemma_token,
)

infile = open("VISA-ESG-txt-spacy.txt","r")
text = infile.read()
#remove \n from file
text = re.sub('\n','',text)
text = re.sub('\t','',text)
text = re.sub('•','',text)
text = re.sub('|','',text)

#text = re.sub(r'\W+','',text)

#convert text into a list
listText = list(text.split(" "))

listText = list(filter(None, listText))
#listText = [i for i in listText if i]


result = pipeline.clean(listText)
result2 = []
for t in result:
    if t !='':
        result2.append(t.lower())

```

```python
#dir(removers)
#dir(spacy_cleaner)
#help(spacy_cleaner.Pipeline)
#help(removers.transformers)
```

```python
#create nlp object
nlp = spacy.load('en_core_web_lg')
# turn result2 list into a str
result2str = ' '.join(result2)
doc = nlp(result2str)
```

```python
from collections import Counter

words = [token.text
         for token in doc
         if not token.is_stop and not token.is_punct and token.is_alpha]

# noun tokens that arent stop words or punctuations
nouns = [token.text
         for token in doc
         if (not token.is_stop and
             not token.is_punct and token.is_alpha and
             token.pos_ == "NOUN")]

# five most common tokens
word_freq = Counter(words)
common_words = word_freq.most_common(400)

# five most common noun tokens
noun_freq = Counter(nouns)
common_nouns = noun_freq.most_common(400)
```

```python

outfile= open("results.txt","w")
for l in common_words:
    a,b = l
    
    outfile.write(str(a)+","+str(b)+"\n")
    
    
outfile.close()

    
```
