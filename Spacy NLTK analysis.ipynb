```python
import spacy
from spacy.tokens import Doc

if not Doc.has_extension("text_id"):
    Doc.set_extension("text_id", default=None)

text_tuples = [
    ("This is the first text.", {"text_id": "text1"}),
    ("This is the second text.", {"text_id": "text2"})
]

nlp = spacy.load("en_core_web_sm")
doc_tuples = nlp.pipe(text_tuples, as_tuples=True)

docs = []
for doc, context in doc_tuples:
    doc._.text_id = context["text_id"]
    docs.append(doc)

for doc in docs:
    print(f"{doc._.text_id}: {doc.text}")
```

```python
import spacy
import spacy_cleaner
from spacy_cleaner.processing import removers, replacers, mutators

model = spacy.load("en_core_web_sm")
pipeline = spacy_cleaner.Pipeline(
    model,
    removers.remove_stopword_token,
    removers.remove_punctuation_token,
    #replacers.replace_punctuation_token,
    mutators.mutate_lemma_token,
)

texts = ["Hello, my name is Cellan! I love to swim!"]

result=pipeline.clean(texts)
# ['hello _IS_PUNCT_ Cellan _IS_PUNCT_ love swim _IS_PUNCT_']
```

```python
import spacy
import spacy_cleaner
from spacy_cleaner.processing import removers, replacers, mutators
from spacy.tokens import Doc
import re

model = spacy.load("en_core_web_lg")
pipeline = spacy_cleaner.Pipeline(
    model,
    removers.remove_stopword_token,
    removers.remove_punctuation_token,
    #replacers.replace_punctuation_token,
    mutators.mutate_lemma_token,
)

#texts = ["Hello, my name is Cellan! I love to swim!"]
#dir(spacy)
nlp = spacy.load('en_core_web_lg')
infile = open("Visa-ESG-txt.txt","r")
text = infile.read()
#remove \n from file
text = re.sub('\n','',text)
text = re.sub('\t','',text)
text = re.sub('•','',text)
text = re.sub('|','',text)

#text = re.sub(r'\W+','',text)

#convert text into a list
listText = list(text.split(" "))

listText = list(filter(None, listText))
#listText = [i for i in listText if i]


result = pipeline.clean(listText)
result2 = []
for t in result:
    if t !='':
        result2.append(t.lower())

```
